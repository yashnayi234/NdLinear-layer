```python
import os
import time
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import psutil

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as transforms

# Import the NdLinear class from the separate file
from ndlinear import NdLinear

# Set random seed for reproducibility
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)
    torch.backends.cudnn.deterministic = True

# Configuration
class Config:
    def __init__(self):
        self.batch_size = 128
        self.epochs = 20  # Train for more epochs to see clearer differences
        self.lr = 0.001
        self.weight_decay = 1e-5  # Adding weight decay
        self.models = ['ndlinear']  # Only using NdLinear model
        self.save_model = False
        self.num_workers = 2  # Adjust based on your system

# Create configuration
config = Config()

# Check for CUDA availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Data transformation
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

# Load FashionMNIST dataset
print("Loading FashionMNIST dataset...")
trainset = torchvision.datasets.FashionMNIST(
    root='./data', train=True, download=True, transform=transform)
trainloader = DataLoader(
    trainset, batch_size=config.batch_size, shuffle=True, num_workers=config.num_workers)

testset = torchvision.datasets.FashionMNIST(
    root='./data', train=False, download=True, transform=transform)
testloader = DataLoader(
    testset, batch_size=config.batch_size, shuffle=False, num_workers=config.num_workers)

# Class labels for FashionMNIST
classes = ('T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
           'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle Boot')

# Show sample images
def show_samples():
    dataiter = iter(trainloader)
    images, labels = next(dataiter)
    
    # Show images
    plt.figure(figsize=(10, 4))
    for i in range(8):
        plt.subplot(2, 4, i+1)
        plt.imshow(images[i].numpy().squeeze(), cmap='gray')
        plt.title(classes[labels[i]])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Display sample images
show_samples()

class NdLinearNet(nn.Module):
    def __init__(self, dropout_rate=0.25):
        super(NdLinearNet, self).__init__()
        
        # Enhanced convolutional layers with batch normalization
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.conv3 = nn.Conv2d(64, 64, kernel_size=3, padding=1)
        self.bn3 = nn.BatchNorm2d(64)
        
        # Pooling and dropout
        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(dropout_rate)
        
        # Sequential NdLinear transformations
        # First NdLinear transforms (64, 7, 7) to (48, 5, 5)
        self.ndlinear1 = NdLinear(
            input_dims=(64, 7, 7), 
            hidden_size=(48, 5, 5),
            transform_outer=True
        )
        
        # Second NdLinear transforms (48, 5, 5) to (32, 4, 4)
        self.ndlinear2 = NdLinear(
            input_dims=(48, 5, 5), 
            hidden_size=(32, 4, 4),
            transform_outer=False  # Alternate transformation order
        )
        
        # Use 2D batch norm instead of 3D
        self.bn_nd = nn.BatchNorm2d(32)
        
        # Final classification layers with reduced parameters
        self.fc1 = nn.Linear(32 * 4 * 4, 128)
        self.bn_fc = nn.BatchNorm1d(128)
        self.fc2 = nn.Linear(128, 10)
    
    def forward(self, x):
        # First conv block
        x = F.relu(self.bn1(self.conv1(x)))
        x = self.pool(x)
        
        # Second conv block
        x = F.relu(self.bn2(self.conv2(x)))
        x = self.pool(F.relu(self.bn3(self.conv3(x))))
        
        # Store original features for residual connection
        original_features = x
        
        # Apply first NdLinear transformation
        x = self.ndlinear1(x)
        
        # Apply second NdLinear transformation
        x = self.ndlinear2(x)
        
        # Normalize and apply non-linearity
        x = F.relu(self.bn_nd(x))
        
        # Flatten for final classification
        x = x.view(x.shape[0], -1)
        
        # Classification head
        x = F.relu(self.bn_fc(self.fc1(x)))
        x = self.dropout(x)
        
        # Final classification layer
        x = self.fc2(x)
        
        return x

# Function to count parameters
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# Detailed parameter breakdown function
def model_parameter_breakdown(model):
    """Print detailed parameter counts for each layer in the model"""
    print("\nDetailed Parameter Breakdown:")
    print("-" * 60)
    print(f"{'Layer':<30} {'Shape':<20} {'Parameters'}")
    print("-" * 60)
    
    total_params = 0
    for name, parameter in model.named_parameters():
        if not parameter.requires_grad:
            continue
        param_count = parameter.numel()
        total_params += param_count
        print(f"{name:<30} {str(list(parameter.shape)):<20} {param_count}")
    
    print("-" * 60)
    print(f"{'Total':<30} {'':<20} {total_params}")
    print("-" * 60)
    return total_params

# Training function
def train(model, criterion, optimizer, epoch):
    model.train()
    train_loss = 0
    correct = 0
    total = 0
    
    # Memory usage tracking
    process = psutil.Process(os.getpid())
    start_memory = process.memory_info().rss / 1024 ** 2  # in MB
    
    start_time = time.time()
    
    with tqdm(trainloader, desc=f"Epoch {epoch}", leave=False) as pbar:
        for batch_idx, (inputs, targets) in enumerate(pbar):
            inputs, targets = inputs.to(device), targets.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()
            
            pbar.set_postfix(
                loss=train_loss/(batch_idx+1),
                acc=100.*correct/total
            )
    
    end_time = time.time()
    end_memory = process.memory_info().rss / 1024 ** 2  # in MB
    memory_usage = end_memory - start_memory
    
    return train_loss/len(trainloader), 100.*correct/total, end_time-start_time, memory_usage

# Evaluation function
def test(model, criterion, epoch):
    model.eval()
    test_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        with tqdm(testloader, desc=f"Validation {epoch}", leave=False) as pbar:
            for batch_idx, (inputs, targets) in enumerate(pbar):
                inputs, targets = inputs.to(device), targets.to(device)
                
                outputs = model(inputs)
                loss = criterion(outputs, targets)
                
                test_loss += loss.item()
                _, predicted = outputs.max(1)
                total += targets.size(0)
                correct += predicted.eq(targets).sum().item()
                
                pbar.set_postfix(
                    loss=test_loss/(batch_idx+1),
                    acc=100.*correct/total
                )
    
    return test_loss/len(testloader), 100.*correct/total

# Main training loop
def run_experiment():
    results = {}
    model_params = {}
    
    for model_type in config.models:
        print(f"\n{'='*50}")
        print(f"Training {model_type.upper()} model")
        print(f"{'='*50}")
        
        # Initialize NdLinear model
        model = NdLinearNet().to(device)
        
        # Print model architecture and parameter count
        print(f"Model Architecture:")
        print(model)
        total_params = count_parameters(model)
        print(f"Total trainable parameters: {total_params:,}")
        
        # Get detailed parameter breakdown
        detailed_params = model_parameter_breakdown(model)
        model_params[model_type] = total_params
        
        # Initialize criterion and optimizer
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=config.lr, weight_decay=config.weight_decay)
        
        # Training variables
        train_losses = []
        test_losses = []
        train_accs = []
        test_accs = []
        times = []
        memory_usages = []
        
        # Training loop
        for epoch in range(1, config.epochs + 1):
            train_loss, train_acc, epoch_time, memory_usage = train(model, criterion, optimizer, epoch)
            test_loss, test_acc = test(model, criterion, epoch)
            
            train_losses.append(train_loss)
            test_losses.append(test_loss)
            train_accs.append(train_acc)
            test_accs.append(test_acc)
            times.append(epoch_time)
            memory_usages.append(memory_usage)
            
            print(f"Epoch {epoch} - Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%, "
                  f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.2f}%, "
                  f"Time: {epoch_time:.2f}s, Memory: {memory_usage:.2f} MB")
        
        # Save results
        results[model_type] = {
            'train_losses': train_losses,
            'test_losses': test_losses,
            'train_accs': train_accs,
            'test_accs': test_accs,
            'times': times,
            'memory_usages': memory_usages,
            'best_accuracy': max(test_accs),
            'avg_time': np.mean(times),
            'avg_memory': np.mean(memory_usages),
            'parameters': model_params[model_type]
        }
        
        # Save model if requested
        if config.save_model:
            model_path = f"fashion_mnist_{model_type}_model.pth"
            torch.save(model.state_dict(), model_path)
            print(f"Model saved to {model_path}")
    
    return results, model_params

# Run the experiment
results, model_params = run_experiment()

# Display results
print("\n\n" + "="*50)
print("RESULTS")
print("="*50)

for model_type in config.models:
    print(f"\n{model_type.upper()} MODEL:")
    print(f"Total parameters: {model_params[model_type]:,}")
    print(f"Best validation accuracy: {results[model_type]['best_accuracy']:.2f}%")
    print(f"Average epoch time: {results[model_type]['avg_time']:.2f}s")
    print(f"Average memory usage: {results[model_type]['avg_memory']:.2f} MB")

# Plot results
def plot_results(results):
    fig, axs = plt.subplots(2, 2, figsize=(15, 12))
    
    # Training Loss
    axs[0, 0].plot(results['ndlinear']['train_losses'], 'r-', label='NdLinear Train')
    axs[0, 0].set_title('Training Loss')
    axs[0, 0].set_xlabel('Epochs')
    axs[0, 0].set_ylabel('Loss')
    axs[0, 0].legend()
    axs[0, 0].grid(True)
    
    # Validation Loss
    axs[0, 1].plot(results['ndlinear']['test_losses'], 'r-', label='NdLinear Val')
    axs[0, 1].set_title('Validation Loss')
    axs[0, 1].set_xlabel('Epochs')
    axs[0, 1].set_ylabel('Loss')
    axs[0, 1].legend()
    axs[0, 1].grid(True)
    
    # Training Accuracy
    axs[1, 0].plot(results['ndlinear']['train_accs'], 'r-', label='NdLinear Train')
    axs[1, 0].set_title('Training Accuracy')
    axs[1, 0].set_xlabel('Epochs')
    axs[1, 0].set_ylabel('Accuracy (%)')
    axs[1, 0].legend()
    axs[1, 0].grid(True)
    
    # Validation Accuracy
    axs[1, 1].plot(results['ndlinear']['test_accs'], 'r-', label='NdLinear Val')
    axs[1, 1].set_title('Validation Accuracy')
    axs[1, 1].set_xlabel('Epochs')
    axs[1, 1].set_ylabel('Accuracy (%)')
    axs[1, 1].legend()
    axs[1, 1].grid(True)
    
    plt.tight_layout()
    plt.savefig('fashion_mnist_ndlinear_performance.png')
    plt.show()
    
    # Additional performance metrics
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    
    # Training Time
    axs[0].plot(results['ndlinear']['times'], 'r-', label='NdLinear')
    axs[0].set_title('Training Time per Epoch')
    axs[0].set_xlabel('Epochs')
    axs[0].set_ylabel('Time (s)')
    axs[0].legend()
    axs[0].grid(True)
    
    # Memory Usage
    axs[1].plot(results['ndlinear']['memory_usages'], 'r-', label='NdLinear')
    axs[1].set_title('Memory Usage per Epoch')
    axs[1].set_xlabel('Epochs')
    axs[1].set_ylabel('Memory (MB)')
    axs[1].legend()
    axs[1].grid(True)
    
    plt.tight_layout()
    plt.savefig('fashion_mnist_ndlinear_resource_usage.png')
    plt.show()

# Plot results
plot_results(results)

# Final conclusions
print("\nConclusions:")
print("="*50)
print(f"NdLinear model: {model_params['ndlinear']:,} parameters")
print(f"Best validation accuracy: {results['ndlinear']['best_accuracy']:.2f}%")
print(f"Average training time per epoch: {results['ndlinear']['avg_time']:.2f}s")
print(f"Average memory usage: {results['ndlinear']['avg_memory']:.2f} MB")

def calculate_convergence(acc_history, best_acc):
    threshold = 0.9 * best_acc
    for i, acc in enumerate(acc_history):
        if acc >= threshold:
            return i + 1
    return len(acc_history)

ndlinear_convergence = calculate_convergence(results['ndlinear']['test_accs'], results['ndlinear']['best_accuracy'])
print(f"Convergence speed: Reached 90% of best accuracy in {ndlinear_convergence} epochs")

import pickle
with open('fashion_mnist_ndlinear_results.pkl', 'wb') as f:
    pickle.dump(results, f)
print("\nResults saved to 'fashion_mnist_ndlinear_results.pkl'")
```
