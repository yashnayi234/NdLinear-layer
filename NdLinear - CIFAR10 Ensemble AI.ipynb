```python
import os
import time
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
import psutil
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as transforms
from ndlinear import NdLinear

# Set random seed for reproducibility
torch.manual_seed(42)
if torch.cuda.is_available():
    torch.cuda.manual_seed(42)
    torch.backends.cudnn.deterministic = True

# Configuration
class Config:
    def __init__(self):
        self.batch_size = 128
        self.epochs = 25  # Train for more epochs to see clearer differences
        self.lr = 0.002
        self.models = ['standard', 'ndlinear']  # Both models to compare
        self.save_model = False
        self.num_workers = 2  # Adjust based on your system

# Create configuration
config = Config()

# Check for CUDA availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Data Augmentation and Normalization
transform_train = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.4914, 0.4822, 0.4465), (0.2023, 0.1994, 0.2010)),
])

# Load CIFAR10 dataset
print("Loading CIFAR10 dataset...")
trainset = torchvision.datasets.CIFAR10(
    root='./data', train=True, download=True, transform=transform_train)
trainloader = DataLoader(
    trainset, batch_size=config.batch_size, shuffle=True, num_workers=config.num_workers)

testset = torchvision.datasets.CIFAR10(
    root='./data', train=False, download=True, transform=transform_test)
testloader = DataLoader(
    testset, batch_size=config.batch_size, shuffle=False, num_workers=config.num_workers)

classes = ('plane', 'car', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck')

# Show sample images
def show_samples():
    dataiter = iter(trainloader)
    images, labels = next(dataiter)
    
    # Show images
    plt.figure(figsize=(10, 4))
    for i in range(8):
        plt.subplot(2, 4, i+1)
        img = images[i].numpy().transpose((1, 2, 0))
        # Unnormalize
        img = img * np.array((0.2023, 0.1994, 0.2010)) + np.array((0.4914, 0.4822, 0.4465))
        img = np.clip(img, 0, 1)
        plt.imshow(img)
        plt.title(classes[labels[i]])
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Display sample images
show_samples()

# Define CNN model with traditional Linear layers
class StandardLinearNet(nn.Module):
    def __init__(self):
        super(StandardLinearNet, self).__init__()
        
        # Convolutional layers
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        
        # Pooling and dropout
        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(0.25)
        
        # Fully connected layers
        self.fc1 = nn.Linear(128 * 4 * 4, 256)
        self.fc2 = nn.Linear(256, 10)
    
    def forward(self, x):
        # Convolutional layers with ReLU and pooling
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        
        # Flatten and pass through fully connected layers
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)
        
        return x

# Define CNN model with NdLinear
class NdLinearNet(nn.Module):
    def __init__(self):
        super(NdLinearNet, self).__init__()
        
        # Convolutional layers
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        
        # Pooling and dropout
        self.pool = nn.MaxPool2d(2, 2)
        self.dropout = nn.Dropout(0.25)
        
        # NdLinear layer to transform (128, 4, 4) to (64, 2, 2)
        self.ndlinear = NdLinear(input_dims=(128, 4, 4), 
                                 hidden_size=(64, 2, 2),
                                 transform_outer=True)
        
        # Final classification layer
        self.fc = nn.Linear(64 * 2 * 2, 10)
    
    def forward(self, x):
        # Convolutional layers with ReLU and pooling
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        
        # Use NdLinear transformation instead of flattening
        x = self.ndlinear(x)
        
        # Flatten the result for the final classification layer
        x = x.view(x.size(0), -1)
        x = self.dropout(x)
        x = self.fc(x)
        
        return x

# Function to count parameters
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

# Detailed parameter breakdown function
def model_parameter_breakdown(model):
    """Print detailed parameter counts for each layer in the model"""
    print("\nDetailed Parameter Breakdown:")
    print("-" * 60)
    print(f"{'Layer':<30} {'Shape':<20} {'Parameters'}")
    print("-" * 60)
    
    total_params = 0
    for name, parameter in model.named_parameters():
        if not parameter.requires_grad:
            continue
        param_count = parameter.numel()
        total_params += param_count
        print(f"{name:<30} {str(list(parameter.shape)):<20} {param_count}")
    
    print("-" * 60)
    print(f"{'Total':<30} {'':<20} {total_params}")
    print("-" * 60)
    return total_params

# Training function
def train(model, criterion, optimizer, epoch):
    model.train()
    train_loss = 0
    correct = 0
    total = 0
    
    # Memory usage tracking
    process = psutil.Process(os.getpid())
    start_memory = process.memory_info().rss / 1024 ** 2  # in MB
    
    start_time = time.time()
    
    with tqdm(trainloader, desc=f"Epoch {epoch}", leave=False) as pbar:
        for batch_idx, (inputs, targets) in enumerate(pbar):
            inputs, targets = inputs.to(device), targets.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            
            train_loss += loss.item()
            _, predicted = outputs.max(1)
            total += targets.size(0)
            correct += predicted.eq(targets).sum().item()
            
            pbar.set_postfix(
                loss=train_loss/(batch_idx+1),
                acc=100.*correct/total
            )
    
    end_time = time.time()
    end_memory = process.memory_info().rss / 1024 ** 2  # in MB
    memory_usage = end_memory - start_memory
    
    return train_loss/len(trainloader), 100.*correct/total, end_time-start_time, memory_usage

# Evaluation function
def test(model, criterion, epoch):
    model.eval()
    test_loss = 0
    correct = 0
    total = 0
    
    with torch.no_grad():
        with tqdm(testloader, desc=f"Validation {epoch}", leave=False) as pbar:
            for batch_idx, (inputs, targets) in enumerate(pbar):
                inputs, targets = inputs.to(device), targets.to(device)
                
                outputs = model(inputs)
                loss = criterion(outputs, targets)
                
                test_loss += loss.item()
                _, predicted = outputs.max(1)
                total += targets.size(0)
                correct += predicted.eq(targets).sum().item()
                
                pbar.set_postfix(
                    loss=test_loss/(batch_idx+1),
                    acc=100.*correct/total
                )
    
    return test_loss/len(testloader), 100.*correct/total

# Main training loop for both models
def run_experiment():
    results = {}
    model_params = {}
    
    for model_type in config.models:
        print(f"\n{'='*50}")
        print(f"Training {model_type.upper()} model")
        print(f"{'='*50}")
        
        # Initialize model based on type
        if model_type == 'standard':
            model = StandardLinearNet().to(device)
        else:
            model = NdLinearNet().to(device)
        
        # Print model architecture and parameter count
        print(f"Model Architecture:")
        print(model)
        total_params = count_parameters(model)
        print(f"Total trainable parameters: {total_params:,}")
        
        # Get detailed parameter breakdown
        detailed_params = model_parameter_breakdown(model)
        model_params[model_type] = total_params
        
        # Initialize criterion and optimizer
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=config.lr)
        
        # Training variables
        train_losses = []
        test_losses = []
        train_accs = []
        test_accs = []
        times = []
        memory_usages = []
        
        # Training loop
        for epoch in range(1, config.epochs + 1):
            train_loss, train_acc, epoch_time, memory_usage = train(model, criterion, optimizer, epoch)
            test_loss, test_acc = test(model, criterion, epoch)
            
            train_losses.append(train_loss)
            test_losses.append(test_loss)
            train_accs.append(train_acc)
            test_accs.append(test_acc)
            times.append(epoch_time)
            memory_usages.append(memory_usage)
            
            print(f"Epoch {epoch} - Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%, "
                  f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.2f}%, "
                  f"Time: {epoch_time:.2f}s, Memory: {memory_usage:.2f} MB")
        
        # Save results
        results[model_type] = {
            'train_losses': train_losses,
            'test_losses': test_losses,
            'train_accs': train_accs,
            'test_accs': test_accs,
            'times': times,
            'memory_usages': memory_usages,
            'best_accuracy': max(test_accs),
            'avg_time': np.mean(times),
            'avg_memory': np.mean(memory_usages),
            'parameters': model_params[model_type]
        }
        
        # Save model if requested
        if config.save_model:
            model_path = f"cifar10_{model_type}_model.pth"
            torch.save(model.state_dict(), model_path)
            print(f"Model saved to {model_path}")
    
    return results, model_params

# Run the experiment
results, model_params = run_experiment()

# Display comparative results
print("\n\n" + "="*50)
print("COMPARISON RESULTS")
print("="*50)

for model_type in config.models:
    print(f"\n{model_type.upper()} MODEL:")
    print(f"Total parameters: {model_params[model_type]:,}")
    print(f"Best validation accuracy: {results[model_type]['best_accuracy']:.2f}%")
    print(f"Average epoch time: {results[model_type]['avg_time']:.2f}s")
    print(f"Average memory usage: {results[model_type]['avg_memory']:.2f} MB")

# Plot comparative results
def plot_comparison(results, model_params):
    fig, axs = plt.subplots(2, 2, figsize=(15, 12))
    
    # Training Loss
    axs[0, 0].plot(results['standard']['train_losses'], 'b-', label='Standard Train')
    axs[0, 0].plot(results['ndlinear']['train_losses'], 'r-', label='NdLinear Train')
    axs[0, 0].set_title('Training Loss')
    axs[0, 0].set_xlabel('Epochs')
    axs[0, 0].set_ylabel('Loss')
    axs[0, 0].legend()
    axs[0, 0].grid(True)
    
    # Validation Loss
    axs[0, 1].plot(results['standard']['test_losses'], 'b-', label='Standard Val')
    axs[0, 1].plot(results['ndlinear']['test_losses'], 'r-', label='NdLinear Val')
    axs[0, 1].set_title('Validation Loss')
    axs[0, 1].set_xlabel('Epochs')
    axs[0, 1].set_ylabel('Loss')
    axs[0, 1].legend()
    axs[0, 1].grid(True)
    
    # Training Accuracy
    axs[1, 0].plot(results['standard']['train_accs'], 'b-', label='Standard Train')
    axs[1, 0].plot(results['ndlinear']['train_accs'], 'r-', label='NdLinear Train')
    axs[1, 0].set_title('Training Accuracy')
    axs[1, 0].set_xlabel('Epochs')
    axs[1, 0].set_ylabel('Accuracy (%)')
    axs[1, 0].legend()
    axs[1, 0].grid(True)
    
    # Validation Accuracy
    axs[1, 1].plot(results['standard']['test_accs'], 'b-', label='Standard Val')
    axs[1, 1].plot(results['ndlinear']['test_accs'], 'r-', label='NdLinear Val')
    axs[1, 1].set_title('Validation Accuracy')
    axs[1, 1].set_xlabel('Epochs')
    axs[1, 1].set_ylabel('Accuracy (%)')
    axs[1, 1].legend()
    axs[1, 1].grid(True)
    
    plt.tight_layout()
    plt.savefig('cifar10_model_comparison.png')
    plt.show()
    
    # Additional performance metrics
    fig, axs = plt.subplots(1, 3, figsize=(18, 6))
    
    # Training Time
    axs[0].plot(results['standard']['times'], 'b-', label='Standard')
    axs[0].plot(results['ndlinear']['times'], 'r-', label='NdLinear')
    axs[0].set_title('Training Time per Epoch')
    axs[0].set_xlabel('Epochs')
    axs[0].set_ylabel('Time (s)')
    axs[0].legend()
    axs[0].grid(True)
    
    # Memory Usage
    axs[1].plot(results['standard']['memory_usages'], 'b-', label='Standard')
    axs[1].plot(results['ndlinear']['memory_usages'], 'r-', label='NdLinear')
    axs[1].set_title('Memory Usage per Epoch')
    axs[1].set_xlabel('Epochs')
    axs[1].set_ylabel('Memory (MB)')
    axs[1].legend()
    axs[1].grid(True)
    
    # Parameter Count
    axs[2].bar(['Standard', 'NdLinear'], [model_params['standard'], model_params['ndlinear']])
    axs[2].set_title('Model Parameter Count')
    axs[2].set_ylabel('Number of Parameters')
    for i, v in enumerate([model_params['standard'], model_params['ndlinear']]):
        axs[2].text(i, v + 0.05, f"{v:,}", ha='center')
    axs[2].grid(True, axis='y')
    
    plt.tight_layout()
    plt.savefig('cifar10_performance_comparison.png')
    plt.show()
    
    # Summary bar chart
    labels = ['Parameters (10)', 'Best Accuracy (%)', 'Avg Time (s)', 'Avg Memory (MB)']
    standard_metrics = [model_params['standard']/1000, 
                      results['standard']['best_accuracy'], 
                      results['standard']['avg_time'], 
                      results['standard']['avg_memory']]
    ndlinear_metrics = [model_params['ndlinear']/1000,
                      results['ndlinear']['best_accuracy'], 
                      results['ndlinear']['avg_time'], 
                      results['ndlinear']['avg_memory']]
    
    x = np.arange(len(labels))
    width = 0.35
    
    fig, ax = plt.subplots(figsize=(12, 6))
    rects1 = ax.bar(x - width/2, standard_metrics, width, label='Standard')
    rects2 = ax.bar(x + width/2, ndlinear_metrics, width, label='NdLinear')
    
    ax.set_title('Performance Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    
    # Add value labels on top of bars
    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate(f'{height:.2f}',
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom')
    
    autolabel(rects1)
    autolabel(rects2)
    
    plt.tight_layout()
    plt.savefig('cifar10_summary_comparison.png')
    plt.show()

# Plot comparison results
plot_comparison(results, model_params)

# Print detailed performance analysis
print("\nPerformance Analysis:")
print("="*50)

# Calculate differences
param_diff = model_params['ndlinear'] - model_params['standard']
param_diff_percent = (param_diff / model_params['standard']) * 100
acc_diff = results['ndlinear']['best_accuracy'] - results['standard']['best_accuracy']
time_diff = results['ndlinear']['avg_time'] - results['standard']['avg_time']
memory_diff = results['ndlinear']['avg_memory'] - results['standard']['avg_memory']

print(f"Parameter Difference: {param_diff:,} ({param_diff_percent:.2f}% {'more' if param_diff > 0 else 'fewer'} for NdLinear)")
print(f"Accuracy Difference: {acc_diff:.2f}% ({'better' if acc_diff > 0 else 'worse'} for NdLinear)")
print(f"Time Difference: {time_diff:.2f}s ({'slower' if time_diff > 0 else 'faster'} for NdLinear)")
print(f"Memory Difference: {memory_diff:.2f} MB ({'more' if memory_diff > 0 else 'less'} for NdLinear)")

# Calculate convergence rate (epochs to reach 90% of best accuracy)
def calculate_convergence(acc_history, best_acc):
    threshold = 0.9 * best_acc
    for i, acc in enumerate(acc_history):
        if acc >= threshold:
            return i + 1
    return len(acc_history)

standard_convergence = calculate_convergence(results['standard']['test_accs'], results['standard']['best_accuracy'])
ndlinear_convergence = calculate_convergence(results['ndlinear']['test_accs'], results['ndlinear']['best_accuracy'])

# Final conclusions
print("\nConclusions:")
print("="*50)
print(f"Standard model: {model_params['standard']:,} parameters")
print(f"NdLinear model: {model_params['ndlinear']:,} parameters")

if param_diff < 0:
    print(f"The NdLinear model uses {-param_diff:,} fewer parameters ({-param_diff_percent:.2f}% reduction).")
else:
    print(f"The NdLinear model uses {param_diff:,} more parameters ({param_diff_percent:.2f}% increase).")

if acc_diff > 0:
    print(f"The NdLinear model achieved {acc_diff:.2f}% better accuracy than the standard model.")
    if ndlinear_convergence < standard_convergence:
        print(f"It also converged faster ({ndlinear_convergence} vs {standard_convergence} epochs).")
    else:
        print(f"However, it took longer to converge ({ndlinear_convergence} vs {standard_convergence} epochs).")
else:
    print(f"The standard model achieved {-acc_diff:.2f}% better accuracy than the NdLinear model.")
    if standard_convergence < ndlinear_convergence:
        print(f"It also converged faster ({standard_convergence} vs {ndlinear_convergence} epochs).")
    else:
        print(f"However, it took longer to converge ({standard_convergence} vs {ndlinear_convergence} epochs).")

if time_diff > 0:
    print(f"The NdLinear model was {time_diff:.2f}s slower per epoch on average.")
else:
    print(f"The NdLinear model was {-time_diff:.2f}s faster per epoch on average.")

if memory_diff > 0:
    print(f"The NdLinear model used {memory_diff:.2f} MB more memory on average.")
else:
    print(f"The NdLinear model used {-memory_diff:.2f} MB less memory on average.")


# Save results dictionary to file
import pickle
with open('cifar10_comparison_results.pkl', 'wb') as f:
    pickle.dump(results, f)
print("\nResults saved to 'cifar10_comparison_results.pkl'")
```
